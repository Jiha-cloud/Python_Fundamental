```python
german_sample3 = german[['Type of apartment','Sex & Marital Status','Credit Amount']]
for type, group in german_sample3.groupby('Type of apartment'):
    print(type)
    print(group.head(n=3))
```

```python
german_sample3
```

```python
for type1, group in german_sample3.groupby('Type of apartment'):
    print('type = ',type1)
    print(group.head(n=3))
```

```python
for type1, group in german_sample3.groupby('Type of apartment'):
    print('type = ',type1)
    print(group.shape)
```

```python
for (type,sex), group in german_sample3.groupby(['Type of apartment',
                                                'Sex & Marital Status']):
    print('type = {}, sex = {}'.format(type,sex))
    print(group.head(n=3))
```

```python
import requests
import urllib.request
from bs4 import BeautifulSoup as bs
```

```python
headers = {'User-Agent':'Mozilla/5.0'}
url = 'https://kissme-wallpapers.com/2101'
html = requests.get(url, headers = headers).text
```

```python
html
```

```python
soup = bs(html, 'html.parser')
articles = soup.find('div',{'class':'tt_article_useless_p_margin contents_style'})
p = articles.find_all('p')
type(p)
```

```python
i=1
for item in p:
    try:
        img = item.find('span',{'class':'imageblock'}).find('img')
        print('img = ',img['src'])
        print('--------------------------')
        urllib.request.urlretrieve(img['src'], 'C:/Users/학생/Desktop/박규영이쁜사진{}.jpg'.format(i))
        i+=1
    except:
        print('No image')
```

```python
area_related = soup.find('div',{'class':'area_related'})
a = area_related.find_all('a')

for item in a:
    txt = item.find('div',{'class':'txt_related'})
    print('https://kissme_wallpapers.com'+item['href'])
    print(txt.text,'\n')
```

```python
area_popular = soup.find('ul',{'id':'popular'})
a = area_popular.find_all('a')

for item in a:
    txt = item.find('div',{'class':'text'})
    print('https://kissme_wallpapers.com'+item['href'])
    print(txt.text,'\n')

```

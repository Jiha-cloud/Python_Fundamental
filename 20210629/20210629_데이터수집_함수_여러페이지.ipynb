```python
import requests
import urllib.request
from bs4 import BeautifulSoup as bs
import pandas as pd
```

```python
# 날짜는 2101-0628, 2021-06-29 2일...
# 4번의 페이지를 수집
# 데이터 프레임을 누적
```

```python
def get_date():
    date = str(input("날짜를 입력하세요 ex)20210629 -> "))
    return date
```

```python
def get_page():
    page = str(input("원하는 페이지를 입력하세요 ex) 3 -> "))
    return page
```

```python
def get_url():
    date = get_date()
    page = get_page()
    url = 'https://news.naver.com/main/list.nhn?mode=LSD&mid=sec&sid1=001&date={}&page={}'.format(date,page)
    lists = [date,page,url]
    return lists
```

```python
def get_html(url):
    headers = {'User-Agent':'Mozilla/5.0'}
    html = requests.get(url, headers = headers).text
    soup = bs(html, 'html.parser')
    type06 = soup.find('ul',{'class':'type06_headline'})
    dl = type06.find_all('dl')
    
    return dl
```

```python
def page_to_new_items():
    img_url = []
    article_url = []
    article_title = []
    article_text = []
    pages = []
    dates = []
    lists = []
    
    lists = get_url()
    dl = get_html(lists[2])
    
    for item in dl:
        try:
            link = item.find('dt',{'class':'photo'})
            img = link.find('img')
            art_u = link.find('a')
            link2 = item.find('dd').find('span',{'class':'lede'})
        
            img_url.append(img['src'])
            article_url.append(art_u['href'])
            article_title.append(img['alt'])
            article_text.append(link2.text)
            pages.append(lists[1])
            dates.append(lists[0])
            
        except:
            continue
    
    article_emg = pd.DataFrame(data = list(zip(article_title,article_url,img_url,article_text,dates,pages)),
                                    columns=['기사제목','기사링크','이미지링크','기사본문','날짜','페이지'])
    pd.set_option('display.max.colwidth', 60)
    
    return article_emg
    
```

```python
def add_articles(article):
    
    lists = []
    lists = get_url()
    dl = get_html(lists[2])

    for item in dl:
        try:
            link = item.find('dt',{'class':'photo'})
            img = link.find('img')
            art_u = link.find('a')
            link2 = item.find('dd').find('span',{'class':'lede'})
            
            article = article.append({'기사제목':img['alt'],'기사링크':art_u['href'],
                                              '이미지링크':img['src'],'기사본문':link2.text,
                                              '날짜':lists[0],'페이지':lists[1]},ignore_index=True)
        except:
            print('안됨')
            continue
    print("after = ", article.shape)  
    return article
```

```python
article_df=page_to_new_items()
article_df
```

```python
add_articles(article_df)
```
